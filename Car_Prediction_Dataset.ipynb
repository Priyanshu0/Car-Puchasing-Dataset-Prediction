```
X = dataset.iloc[:,3:-1]
Y = dataset.iloc[:,-1]
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```
from sklearn import linear_model
```

```
regr = linear_model.LinearRegression()
```

```
regr.fit(X_train,y_train)
```

```
predict_train = regr.predict(X_train)
```

```
predict_test = regr.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error, r2_score
```

```
print(f"Training error: {mean_squared_error(predict_train,y_train)}")
print(f"Testing error: {mean_squared_error(predict_test,y_test)}")
```

```
print(f"Training r2_score: {r2_score(predict_train,y_train)}")
print(f"Testing r2_score: {r2_score(predict_test,y_test)}")
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```
normalizer = preprocessing.Normalization()
```

```
normalizer.adapt(np.array(X_train))
```

```
model = keras.Sequential(
    [normalizer,
     layers.Dense(64,activation='relu'),
     layers.Dense(64,activation = 'relu'),
     layers.Dense(1)]
)
```

```
model.compile(optimizer= tf.keras.optimizers.Adam(0.001), loss = 'mean_absolute_error')
```

```
model.summary()
```

```
history = model.fit(X,Y, validation_split = 0.2, verbose = 1, epochs = 1000)
```

```
plt.plot(history.history['loss'],label = 'loss')
plt.plot(history.history['val_loss'],label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.grid()
plt.legend()
plt.show()
```
